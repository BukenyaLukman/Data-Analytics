```python
cars = pd.read_csv('mtcars.csv')
```

```python
cars.columns = ['cars_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']
cars.head()
```

```python
sb.pairplot(cars)
```

```python
x = cars[['mpg','hp','qsec','wt']]
sb.pairplot(x)
```

```python
mpg = cars['mpg']
hp = cars['hp']
qsec = cars['qsec']
wt = cars['wt']
pearsonr_coefficient, p_value = pearsonr(mpg,hp)
print('PearsonR Correlation Coefficent %0.3f',pearsonr_coefficient)
```

```python
pearsonr_coefficient, p_value = pearsonr(mpg,qsec)
print('PearsonR Correlation Coefficent %0.3f',pearsonr_coefficient)
```

```python
pearsonr_coefficient, p_value = pearsonr(mpg,wt)
print('PearsonR Correlation Coefficent %0.3f',pearsonr_coefficient)
```
